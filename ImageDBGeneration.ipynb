```python
cnnArchive = ZipFile("/content/data/cnn_val.zip", 'r')



def getList(archive, subFolderPath):
  file_list = []
  count = 0
  for filename in archive.namelist():
    if str(filename).startswith(subFolderPath):
      file_list.append(filename)
      count = count + 1
      if count >= 500:
        print('Completing the list')
        break;
  print("Number of elements selected", len(file_list))
  return file_list
```

```python
person_List = getList(cnnArchive, "person/0_real")
```

```python
dog_list = getList("dog/0_real")
```

```python
car_List = getList("car/0_real")
```
